## Data Preprocessing

```python
data = pd.read_csv("data/S&P 500 Historical Data (US).csv", index_col='Date',parse_dates=True, dayfirst=False)
begin = "2010-01-01"
end = "2020-01-01"
data_index = data.index
data.index = data.index.to_period(freq="D") # 1(Day)Ïùº
series_list = []
```
